<b>I observed that solidity will not help me to classify a pumpkin</b>

```python
train_df = df.drop(["Solidity", "Class"], axis=1)
test_df = df["Class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_df, test_df, test_size=0.25)
len(x_train), len(y_train)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
